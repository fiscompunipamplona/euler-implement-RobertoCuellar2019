```python
import numpy as np
import matplotlib.pyplot as plt
from EULER import * ##En la clase euler se encuentran los códigos para
# el método de euler y euler mejorado
```

Ejemplo de implementación de método de euler para la ecuación diferencial ordinaria de primer orden $y'=xy$. 

```python
## Se definen las condiciones iniciales, parametros y el respectivo dominio
#dominio___
lim_sup = 10 
lim_inf = 0
#resolución___
h = 0.1
#condicion inicial ___
yo = 3
f_xy = 'x*y' # la funcion debe estar en terminos de x e y para que los
# metodos de las clases, sean correctamente interpretados


evaluar = euler_normal() ## se llama la clase  del algoritmo de euler estandar
## se envia y', lim_sup, lim_inf, h(intervalo), yo(valor inicial)
x_y = np.array(evaluar.euler_1er(f_xy,lim_sup, lim_inf, h, yo)) #el np.array permite  manipular de forma mas sencilla
x = x_y[0,:] #se extrae el vector del dominio
y = x_y[1,:] #se extrae la aproximación hecha
y_teo = yo*np.exp((x**2)/2) # solución analítica a la edo


evaluar_1 = euler_mejorado()### clase algoritmo euler mejorado
x_y_1 = np.array(evaluar_1.euler_1er_m(f_xy,lim_sup, lim_inf, h, yo))## se envian los datos de igual forma al euler estandar
x1 = x_y_1[0,:]
y1 = x_y_1[1,:]
err1_e = abs(y_teo-y) #diferencias relativas
err1_em = abs(y_teo-y1)

ax1 = plt.subplot(2, 1, 1) #gráficas comparativas h =0.1
fig_num, = plt.plot(x,y,label='Euler')
fig_num_1, = plt.plot(x1,y1,label='Euler Mejorado')
fig_teo, =plt.plot(x, y_teo,label='Teorico')
plt.legend(handles=[fig_num,fig_num_1, fig_teo])
plt.ylabel('y')
plt.grid('true')
plt.text(6, 1e22, r'$h$ = %0.1f '%(h), fontsize=10)
plt.setp(ax1.get_xticklabels(), visible=False)



plt.subplot(2, 1, 2) ## para h = 0.05
h = 0.005
evaluar = euler_normal() ## se llama la clase  del algoritmo de euler estandar
## se envia y', lim_sup, lim_inf, h(intervalo), yo(valor inicial)
x_y1 = np.array(evaluar.euler_1er(f_xy,lim_sup, lim_inf, h, yo)) #el np.array permite  manipular de forma mas sencilla
x1 = x_y1[0,:] #se extrae el vector del dominio
y1 = x_y1[1,:] #se extrae la aproximación hecha
y_teo = yo*np.exp((x1**2)/2) # solución analítica a la edo


evaluar_1 = euler_mejorado()### clase algoritmo euler mejorado
x_y_11 = np.array(evaluar_1.euler_1er_m(f_xy,lim_sup, lim_inf, h, yo))## se envian los datos de igual forma al euler estandar
x11 = x_y_11[0,:]
y11 = x_y_11[1,:]

err2_e = abs(y_teo-y1) # diferencias relativas
err2_em = abs(y_teo-y11)

fig_num, = plt.plot(x1,y1,label='Euler')
fig_num_1, = plt.plot(x11,y11,label='Euler Mejorado')
fig_teo, =plt.plot(x1, y_teo,label='Teorico')
plt.legend(handles=[fig_num,fig_num_1, fig_teo])
plt.xlabel('x')
plt.ylabel('y')
plt.grid('on')
plt.text(6, 1e22, r'$h$ = %0.3f '%(h), fontsize=10)
plt.show()

```

```python
ax1 = plt.subplot(2, 1, 1) #gráficas comparativas h =0.1
fig_nume, = plt.plot(x,err1_e,label='Diferencia euler')
fig_num_1e, = plt.plot(x,err1_em,label='Diferencia Euler Mejorado')
plt.legend(handles=[fig_nume,fig_num_1e])
plt.ylabel('Diferencia relativa')
plt.grid('true')
h=0.1
plt.text(6, 1e22, r'$h$ = %0.1f '%(h), fontsize=10)
plt.setp(ax1.get_xticklabels(), visible=False)

plt.subplot(2, 1, 2) ## para h = 0.05
h = 0.005
fig_num, = plt.plot(x1,err2_e,label='Diferencia euler')
fig_num_1, = plt.plot(x1,err2_em,label='Diferencia Euler Mejorado')
plt.legend(handles=[fig_num,fig_num_1])
plt.ylabel('Diferencia relativa')
plt.grid('true')
plt.text(6, 0.5e22, r'$h$ = %0.3f '%(h), fontsize=10)
plt.show()
```

Para una ecuación diferencial de segundo orden $y''+w_{0}^{2}=0$, se tiene lo siguiente

```python
wo = 2
vo = 0
yo = 10
tmax = 10
tmin = 0
h = 0.1
f_xy = '-(2**2)*y'
evaluar = euler_normal()## algoritmo de euler normal
x_y = np.array(evaluar.euler_2do(f_xy,tmax,tmin,h,yo,vo))
t = x_y[0,:]
y = x_y[1,:]
v = x_y[2,:]
y_teo = yo*np.cos(wo*t)

evaluar_1 = euler_mejorado() ## algoritmo de euler mejorado
x_y_1 = np.array(evaluar_1.euler_2do_m(f_xy,tmax,tmin,h,yo,vo))
t1 = x_y_1[0,:]
y1 = x_y_1[1,:]
v1 = x_y_1[2,:]

evaluar_2 = euler_mejorado() ## algoritmo de euler mejorado
x_y_2 = np.array(evaluar_1.euler_2do_cromer(f_xy,tmax,tmin,h,yo,vo))
t2 = x_y_2[0,:]
y2 = x_y_2[1,:]
v2 = x_y_2[2,:]



fig_num, = plt.plot(t ,y,label='Euler')
fig_num_1, = plt.plot(t1 ,y1,label='Euler Mejorado')
fig_num_2, = plt.plot(t2,y2, label='Euler Cromer')
fig_teo, =plt.plot(t, y_teo,label='Teorico')
plt.text(4, 40, r'$h$ = %0.3f '%(h), fontsize=10)
plt.legend(handles=[fig_num,fig_num_1, fig_num_2,fig_teo])
plt.grid('true')
plt.show()
```

Diferencias relativas entre los métodos

```python
err_2de = abs(y_teo- y)
err_2dem = abs(y_teo- y1)
err_2dec = abs(y_teo- y2)

fig_numm, = plt.plot(t ,err_2de,'--',label='Euler')
fig_num_1m, = plt.plot(t1 ,err_2dem,'--',label='Euler Mejorado')
fig_num_2m, = plt.plot(t2,err_2dec,'--', label='Euler Cromer')

plt.text(4, 40, r'$h$ = %0.3f '%(h), fontsize=10)
plt.legend(handles=[fig_numm,fig_num_1m, fig_num_2m])
plt.grid('true')
plt.show()
```
